# Random Forest Regressor

```python
import sklearn
import pandas as pd
```

```python
data = pd.read_csv("D:/Datasets/Machine Learning/tip.csv")
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
data['Gender'] = encoder.fit_transform(data['sex'])
data['day'] = encoder.fit_transform(data['day'])
data['smoker'] = encoder.fit_transform(data['smoker'])
data['time'] = encoder.fit_transform(data['time'])
```

```python
data.head()
```

```python
data.drop(columns='sex',inplace=True)
```

```python
data.info()
```

### Random Forest Uses Decision Tree In The Backend

```python
X = data.drop(columns='total_bill')
y = data['total_bill']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.14)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(r2_score(y_pred,y_test))
```

```python
print(mean_squared_error(y_pred,y_test))
```

##### Booster

```python
from sklearn.ensemble import RandomForestRegressor
booster = RandomForestRegressor(n_estimators=10)
booster.fit(X_train,y_train)
```

```python
y_pred2 = booster.predict(X_test)
```

```python
print(r2_score(y_pred2,y_test))
```
