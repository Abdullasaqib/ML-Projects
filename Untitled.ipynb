```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data  # features (sepal length, sepal width, petal length, petal width)
y = iris.target  # actual labels (setosa, versicolor, virginica)

# Standardize the features for better clustering performance
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
y_kmeans = kmeans.predict(X_scaled)

# Visualizing clusters using two features (sepal length and sepal width for simplicity)
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans, cmap='viridis', s=50)

# Plotting the centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.75, marker='X')

plt.title("K-Means Clustering on Iris Dataset With all Sepal features")
plt.xlabel("Standardized Sepal Length")
plt.ylabel("Standardized Sepal Width")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 2], X_scaled[:, 3], c=y_kmeans, cmap='viridis', s=50)

# Plotting the centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 2], centroids[:, 3], c='red', s=200, alpha=0.75, marker='X')

plt.title("K-Means Clustering on Iris Dataset With all Petal features")
plt.xlabel("Standardized Petal Length")
plt.ylabel("Standardized Petal Width")
plt.grid(True)
plt.show()
```
