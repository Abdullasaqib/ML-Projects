# Support Vector

```python
import sklearn
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Datasets\Machine Learning\titanic.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns='Name',inplace=True)
from sklearn.preprocessing import LabelEncoder
Encoder = LabelEncoder()
df['Gender'] = Encoder.fit_transform(df['Gender'])
```

```python
df.info()
```

```python
X = df.drop(columns='Survived')
y = df['Survived']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

```python
from sklearn.svm import SVC
vector_classifier = SVC()
```

```python
vector_classifier.fit(X_train,y_train)
```

```python
y_pred = vector_classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred, y_test))
```

# K Nearest Neighbor

```python
data = pd.read_csv(r"D:\Datasets\Machine Learning\tip.csv")
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
Encoder = LabelEncoder()
```

```python
data['Gender'] = Encoder.fit_transform(data['sex'])
data.drop(columns='sex',inplace=True)
data['smoker'] = Encoder.fit_transform(data['smoker'])
data['time'] = Encoder.fit_transform(data['time'])
data['day'] = Encoder.fit_transform(data['day'])
```

```python
data.info()
```

```python
A = data.drop(columns='day')
b = data['day']
```

```python
from sklearn.model_selection import train_test_split
A_train, A_test, b_train, b_test = train_test_split(A,b, test_size=0.12,
                                                   random_state=2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neighbor_model = KNeighborsClassifier()
```

```python
neighbor_model.fit(A_train,b_train)
```

```python
b_pred = neighbor_model.predict(A_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(b_pred,b_test))
print(confusion_matrix(b_pred, b_test))
```
