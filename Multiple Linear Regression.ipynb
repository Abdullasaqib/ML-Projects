```python
# Step 1: Importing The Libraries
import pandas as pd
import sklearn
```

```python
# Step 2: Importing The Data
df = pd.read_csv(r"D:\Datasets\Machine Learning\USA_Housing.csv")
```

```python
df.head()
```

```python
# Step 3: Data Preprocessing/ Data Cleaning
df.info()
```

```python
df.drop(columns = 'Address',inplace=True)
```

```python
df['Price'] = df['Price'].astype('int')
```

```python
df.info()
```

#### Step 4: Splitting The Data

```python
# X = df[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms','Area Population']]
X = df.drop(columns='Price') # Features
y = df['Price'] # Target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,
                                                    random_state=123)
```

```python
# Step 5: Building The Model
    # 5.1: Initializing The Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# 5.2: Training The Model
model.fit(X_train, y_train)
```

```python
# 5.3: Predicting From The Model
y_pred = model.predict(X_test)
```

```python
# 5.4: Testing The Model
from sklearn.metrics import r2_score, mean_squared_error
print(r2_score(y_pred,y_test)*100)
print(mean_squared_error(y_pred,y_test)**0.5)
```
