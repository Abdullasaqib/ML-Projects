```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv("D:/Datasets/Machine Learning/USA_Housing.csv")
```

```python
df
```

```python
df.drop(columns='Address',inplace=True)
```

```python
X = df[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Area Population']]
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=123)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
X_train_ = poly.fit_transform(X_train)
X_test_ = poly.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
poly_model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(r2_score(y_pred,y_test)*100)
print(mean_squared_error(y_pred,y_test)**0.5)
```

```python
poly_model.fit(X_train_,y_train)
```

```python
y_pred_ = poly_model.predict(X_test_)
```

```python
print(r2_score(y_pred_,y_test)*100)
print(mean_squared_error(y_pred_,y_test)**0.5)
```
