```python
import pandas as pd
import sklearn
```

```python
data = pd.read_csv("D:/Datasets/Machine Learning/tip.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['sex'] = encoder.fit_transform(data['sex'])
data['smoker'] = encoder.fit_transform(data['smoker'])
data['day'] = encoder.fit_transform(data['day'])
data['time'] = encoder.fit_transform(data['time'])
```

```python
data.info()
```

```python
data['size'].unique()
```

```python
X = data.drop(columns='sex')
y = data['sex']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=741)
```

```python
from sklearn.linear_model import LogisticRegression
cls_mod = LogisticRegression()
```

```python
cls_mod.fit(X_train,y_train)
```

```python
y_pred = cls_mod.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```
