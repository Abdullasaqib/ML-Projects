# Regression

#### Step 1: Importing The Libraries

```python
import sklearn
import pandas as pd
import matplotlib.pyplot as plt
```

#### Step 2: Importing The Data

```python
data = pd.read_csv(r"D:\Datasets\Machine Learning\tip.csv")
```

#### Step 3: Data Cleaning / Data Preprocessing

```python
data.info()
```

```python
plt.scatter(data['total_bill'],data['tip'])
```

#### Step 4: Splitting The Data

```python
A = data[['total_bill']] # Features
b = data['tip'] # Target
```

```python
from sklearn.model_selection import train_test_split
A_train, A_test, b_train, b_test = train_test_split(A,b,test_size=0.2,
                                                   random_state=0)
```

#### Step 5: Building The Model
##### 5.1: Initializing The Model

```python
from sklearn.neighbors import KNeighborsRegressor
neighbor_model = KNeighborsRegressor(n_neighbors=6)
```

##### Step 5.2: Model Training

```python
neighbor_model.fit(A_train,b_train)
```

##### Step 5.3: Predicting From The Model

```python
b_pred = neighbor_model.predict(A_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(r2_score(b_test,b_pred)*100)
print(mean_squared_error(b_test,b_pred)**0.5)
```
