```python
# Step 1: Installing & Importing The Libraries
import sklearn
import pandas as pd
```

```python
# Step 2: Importing The Data
from sklearn.datasets import load_iris
df = load_iris()
```

### This Is A Pre-Defined Data Of Library Scikit-Learn

#### So We Don't Need To Do Data Cleaning/ Pre-Processing On It

```python
# Step 4: Splitting The Data
X = df.data
y = df.target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.21,random_state=753)
```

```python
# Step 5: Building The Model
# 5.1:Initializing The Model
from sklearn.neighbors import KNeighborsClassifier
neighbor_mdl = KNeighborsClassifier()
```

```python
neighbor_mdl.fit(X_train,y_train)
```

```python
y_pred = neighbor_mdl.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```
