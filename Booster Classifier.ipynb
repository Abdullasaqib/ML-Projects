# Random Forest Classifier

```python
import pandas as pd
import sklearn
```

```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
X = data.data
y = data.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.15,random_state=85)
```

```python
from sklearn.ensemble import RandomForestClassifier
booster = RandomForestClassifier(n_estimators=25)
```

```python
booster.fit(X_train,y_train)
```

```python
y_pred = booster.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

# Ada Boost

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.ensemble import AdaBoostClassifier
booster2 = AdaBoostClassifier(estimator=model,n_estimators=85)
```

```python
booster2.fit(X_train,y_train)
```

```python
y_pred2 = booster2.predict(X_test)
```

```python
print(classification_report(y_pred2,y_test))
print(confusion_matrix(y_pred2,y_test))
```
