```python
import pandas as pd
import sklearn
```

```python
df = pd.read_csv(r"D:\Datasets\Machine Learning\USA_Housing.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop(columns='Address',inplace=True)
```

```python
df['Price'] = df['Price'].astype('int')
```

```python
df.info()
```

```python
df.head()
```

```python
X = df.drop(columns='Price')
y = df['Price']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.18, random_state=12)
```

```python
from sklearn.svm import SVR
model = SVR(kernel='linear')
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(r2_score(y_test,y_pred)*100)
print(mean_squared_error(y_test, y_pred)**0.5)
```

## Classification

```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
print(data.DESCR)
```

```python
dataframe = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
dataframe
```

```python
dataframe['Target'] = data.target
```

```python
dataframe
```

```python
A = data.data
b = data.target
A_train, A_test, b_train, b_test = train_test_split(A,b,test_size=0.2,random_state=0)
```

```python
from sklearn.svm import SVC
model2 = SVC()
```

```python
model2.fit(A_train,b_train)
```

```python
b_pred = model2.predict(A_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(b_test,b_pred))
print(confusion_matrix(b_test, b_pred))
```
