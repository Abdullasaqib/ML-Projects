# Regression

### Step 1: Importing The Libraries

```python
import pandas as pd
import sklearn
import seaborn as sns
```

### Step 2: Importing the Data

```python
df = sns.load_dataset('mpg')
```

### Step 3: Data Cleaning / Data Preprocessing

```python
df.info()
```

```python
df.head()
```

```python
df['origin'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
Encoder = LabelEncoder()
df['origin'] = Encoder.fit_transform(df['origin'])
```

```python
df.head()
```

```python
df['horsepower'] = df['horsepower'].fillna(104.4)
```

```python
df.drop(columns='name',inplace=True)
```

```python
df.info()
```

```python
df.corr()
```

### Step 4: Spliting The Data

```python
feature1 = df.drop(columns='mpg')
target1 = df['mpg']
from sklearn.model_selection import train_test_split
feature1_train, feature1_test, target1_train, target1_test = train_test_split(feature1, target1, test_size=0.2,
                                                                             random_state=159)
```

### Step 5: Building The Model

```python
from sklearn.tree import DecisionTreeRegressor
tree_model = DecisionTreeRegressor()
```

```python
tree_model.fit(feature1_train,target1_train)
```

```python
target1_pred = tree_model.predict(feature1_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(r2_score(target1_test,target1_pred)*100)
print(mean_squared_error(target1_test,target1_pred)**0.5)
```

# Classification

```python
feature2 = df.drop(columns='origin')
target2 = df['origin']
```

```python
feature2_train, feature2_test, target2_train, target2_test = train_test_split(feature2,target2,test_size=0.2,
                                                                             random_state=159)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model2 = DecisionTreeClassifier()
```

```python
tree_model2.fit(feature2_train,target2_train)
```

```python
target2_pred = tree_model2.predict(feature2_test)
```

## Testing 

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(target2_test,target2_pred))
print(confusion_matrix(target2_test,target2_pred))
```
