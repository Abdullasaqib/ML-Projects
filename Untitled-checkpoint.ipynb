```python
import sklearn
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Datasets\Machine Learning\student_performance_prediction.csv")
```

```python
df.info()
```

```python
df.drop(columns='Student ID', inplace=True)
df['Study Hours per Week'] = df['Study Hours per Week'].fillna(9.9)
df['Previous Grades'] = df['Previous Grades'].fillna(65.4)
df['Attendance Rate'] = df['Attendance Rate'].fillna(75.2)
df['Participation in Extracurricular Activities'] = df['Participation in Extracurricular Activities'].fillna('No')
df['Parent Education Level'] = df['Parent Education Level'].fillna('Bachelor')
df['Passed'] = df['Passed'].fillna('Yes')
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Participation in Extracurricular Activities'] = encoder.fit_transform(df['Participation in Extracurricular Activities'])
df['Parent Education Level'] = encoder.fit_transform(df['Parent Education Level'])
df['Passed'] = encoder.fit_transform(df['Passed'])
```

```python
df.info()
```

```python
X = df.drop(columns='Passed')
y = df['Passed']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
log_model.fit(X_train,y_train)
```

```python
y_pred = log_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
