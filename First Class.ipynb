# Regression

```python
# Step 1: Installing And Importing Libraries
import sklearn # for Working With Machine Learning Algorithms
import pandas as pd # For Working With Data
import matplotlib.pyplot as plt # For Understanding (or) Visualizing The Data
```

```python
# Step 2: Importing The Data
my_dict = {'Exp': [2,6,4,8,9,3,1,5,7,10],'Sal': [2000,6000,4000,8000,9000,3000,1000,5000,7000,10000]}  
```

```python
# Step 3: Data Cleaning/ Pre-Processing
# We Will Not Do The STep For Now Cause We have Created Our Own Data
```

```python
# Step 4: Splitting The Data
df = pd.DataFrame(my_dict)
```

```python
df
```

```python
X = df[['Exp']] # features
y = df['Sal'] # Target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X , y ,test_size=0.2)
```

```python
# Step 5: Building The Model
    # 5.1: Initializing The Model
from sklearn.linear_model import LinearRegression
first_model = LinearRegression()
```

```python
# Step 5.2: Training The Model
first_model.fit(X_train, y_train)
```

```python
# Step 5.3: Predicting From The Model
y_pred = first_model.predict(X_test)
```

```python
# Step 5.4: Testing The Model
from sklearn.metrics import r2_score, mean_squared_error
print(r2_score(y_pred,y_test))
print(mean_squared_error(y_pred,y_test))
```
