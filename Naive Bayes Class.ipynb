```python
# Step 1: Installing And Importing Libraries
import pandas as pd
import sklearn
```

```python
# Step 2: Importing The Data
df = pd.read_csv("D:/Datasets/Machine Learning/spam.csv", encoding= 'ISO-8859-1')
```

```python
df.head()
```

```python
# Naive Bayes Is An Algorithm In Machine Learning 
# WHich Is The Only ALgorithm That Works With ALphabetical Values
# Spam = Useless Emails
# Ham = Unoffical or Personal And Useful Emails
```

```python
df.info()
```

```python
df.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'],inplace=True)
```

```python
df.info()
```

```python
df.rename(columns={'v1': 'Target','v2':'Messages'},inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Target'] = encoder.fit_transform(df['Target'])
```

```python
df.head()
```

```python
X = df['Messages']
y = df['Target']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.15,random_state=862)
```

```python
# Vectorization Saves The Alphabets With Their Meanings In Numerical Form 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vector = CountVectorizer()
```

```python
X_train_transformed = vector.fit_transform(X_train)
X_test_test = vector.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes_model = MultinomialNB()
```

```python
naive_bayes_model.fit(X_train_transformed,y_train)
```

```python
y_pred = naive_bayes_model.predict(X_test_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

```python
import re
def textcleaning(text):
    text = text.lower()  # For Making It In Lower Case
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'\W+', ' ', text)  # Remove special characters
    text = text.strip()
    return text

def prediction(email):
    cleaned_email = textcleaning(email)
    vectorized_email = vector.transform([cleaned_email])
    prediction = naive_bayes_model.predict(vectorized_email)
    return 'spam' if prediction[0] == 1 else 'ham'
```

```python
email_content = "Congragulations! Buddy You Have Won The Ticket For A World Tour Claim Now"
result = prediction(email_content)
print(f"The email content is classified as: {result}")
```
